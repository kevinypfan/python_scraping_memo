```python
import requests
from bs4 import BeautifulSoup
import re
pages = set()
titles = []
dict_vedio = []
```

```python
response = requests.get('https://www.youtube.com/results?search_query=violeta')
```

```python
html = response.text
```

```python
bs = BeautifulSoup(html, 'html.parser')
print(bs.prettify())
```

```python
all_video_tag = bs.find_all('a', {'href':re.compile('(\/watch)')})
# video_views_list = re.findall('\d+', all_video_tag[5].find('span', {'class': 'view-count'}).text)
# video_views_int = ''.join(video_views_list)
# video_views_int
# dir(all_video_tag[5])
# print(all_video_tag[4].span)
print(all_video_tag[5].span.attrs['aria-label'])
```

```python
for link in all_video_tag:
    if 'href' in link.attrs:
        if link.attrs['href'] not in pages:
            try:
                pages.add(link.attrs['title'])
                titles.append({'title':link.attrs['title'], 'link_id':link.attrs['href'].split('=')[-1], 'views': int(''.join(re.search('([\d,]+)次$', link.span.attrs['aria-label']).group(1).split(','))) })
            except:
                # Key is not present
                pass
            
```
