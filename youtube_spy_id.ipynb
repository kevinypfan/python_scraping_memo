```python
import requests
from bs4 import BeautifulSoup
import re
titles = []
```

```python
response = requests.get('https://www.youtube.com/watch?v=6eEZ7DJMzuk')
```

```python
html = response.text
```

```python
bs = BeautifulSoup(html, 'html.parser')
print(bs.prettify())
```

```python
all_video_tag = bs.find_all('a', {'href': re.compile('(\/watch)'), 'class': re.compile('content-link')})
```

```python
for link in all_video_tag:
    if 'href' in link.attrs:
        try:
            titles.append({'title':link.attrs['title'], 'link_id':link.attrs['href'].split('=')[-1], 'views': int(''.join(re.search('([\d,]+)次$', link.find('span', {'class': re.compile('view-count')}).text).group(1).split(','))) })
        except:
            # Key is not present
            pass
            
```
